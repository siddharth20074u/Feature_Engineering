# Missing Values

```python
dataset.isnull().sum()
```

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
```

```python
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(),4), '%missing values')
```

## Relationship with dependent feature

```python
for feature in features_with_na:
    data = dataset.copy()
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar(color='blue')
    plt.title(feature)
    plt.show()
```

# Numerical values

```python
numerical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes != 'object']
```

```python
print('Number of numerical variables: {}'.format(len(numerical_feature)))
```

```python
dataset[numerical_feature].head()
```

### Temporal variables(Eg: Date/Time)

```python
year_feature = [feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot(c = 'red')
plt.xlabel(feature)
plt.ylabel('median house price')
plt.title('House price vs Yr sold')
```

```python
for feature in year_feature:
    data = dataset.copy()
    data[feature] = data['YrSold']-data[feature]
    plt.scatter(data[feature],data['SalePrice'], c = 'navy')
    plt.xlabel(feature)
    plt.ylabel('Sale price')
    plt.show()
```

## Discrete Feature

```python
discrete_feature = [feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in year_feature + ['Id']]
```

```python
print('Discrete variables count : {}'.format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar(color ='r')
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

## Continous variable

```python
continous_feature = [feature for feature in numerical_feature if feature not in discrete_feature + ['Id'] + year_feature]
print('Continous feature count: {}'.format(len(continous_feature)))
```

```python
for feature in continous_feature:
    data = dataset.copy()
    data[feature].hist(bins = 25, color = 'black')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

## Using logarithmic tranformation

```python
for feature in continous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

### Outliers

```python
for feature in continous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
